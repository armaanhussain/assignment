```python
Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.
Ans1. Matplotlib is a popular Python library used for creating static, interactive, and publication-quality visualizations. It provides a wide variety of plots and charts to represent data in a visually appealing manner.

The Pyplot module within Matplotlib offers a MATLAB-like interface for creating plots. Here are five types of plots that can be created using Pyplot:

1.Line Plot: Useful for displaying trends or relationships between variables by connecting data points with straight lines.
2.Scatter Plot: Shows the relationship between two variables by plotting individual data points on a 2D plane.
3.Bar Plot: Represents categorical data with rectangular bars. Useful for comparing different categories.
4.Histogram: Displays the distribution of numerical data by dividing it into bins and showing the frequency of observations in each bin.
5.Pie Chart: Represents data as a circular statistical graphic divided into slices to illustrate numerical proportions.
```

```python
"""Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.

Ans2 : In Matplotlib, a scatter plot is a type of plot used to display the relationship between two sets of data.
It represents individual data points on a two-dimensional plane, with one variable plotted on the x-axis and another on the y-axis."""

import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()

```

```python
"""Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])"""

#Ans3. The subplot() function in Matplotlib is used to create multiple plots within a single figure. It allows you to arrange
# and display several plots (subplots) in a grid-like fashion, specifying the number of rows, columns, and the position of each subplot.

import matplotlib.pyplot as plt
import numpy as np


# Data for four lines
x = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Creating subplots with 2 rows and 2 columns

# Subplot 1: Line 1
plt.subplot(2, 2, 1)
plt.plot(x, y1, 'r-')
plt.title('Line 1')

# Subplot 2: Line 2
plt.subplot(2, 2, 2)
plt.plot(x, y2, 'g--')
plt.title('Line 2')

# Subplot 3: Line 3
plt.subplot(2, 2, 3)
plt.plot(x, y3, 'b-.')
plt.title('Line 3')

# Subplot 4: Line 4
plt.subplot(2, 2, 4)
plt.plot(x, y4, 'y:')
plt.title('Line 4')

# Adjust layout for better spacing
plt.tight_layout()

# Show the plots
plt.show()
```

```python
#Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
"""Ans4. The bar plot in Matplotlib is used to visually represent categorical data using rectangular bars. It's an effective way to compare data
across different categories, showing the relationships between different groups or entities.
Here are a few reasons why bar plots are used:

1.Comparison: They allow for a straightforward comparison between categories or groups by visually representing their values.

2.Categorical Data Display: Bar plots are effective for displaying categorical data where each bar represents a category and its
length or height corresponds to a particular value or frequency.
"""
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(8,5))
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)

plt.figure(figsize=(8,5))
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.barh(company,profit)

```

```python
"""Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
Ans5. A box plot, also known as a box-and-whisker plot, is a graphical representation that displays the distribution of a dataset
along with its key statistical measures such as median, quartiles, and outliers. It's particularly useful for visualizing the spread
and skewness of the data, as well as identifying outliers.
"""
import matplotlib.pyplot as plt
import numpy as np

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]

# Creating a box plot
plt.figure(figsize=(8, 6))
plt.boxplot(data)
plt.xticks([1, 2], ['Box 1', 'Box 2'])
plt.xlabel('Box')
plt.ylabel('Values')
plt.title('Box Plot of Box 1 and Box 2')
plt.grid(True)
plt.show()

```
