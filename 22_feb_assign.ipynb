```python
Q5. What are some common functions you can use to manipulate data in a Pandas DataFrame? Can
you give an example of when you might use one of these functions?

Ans5. Pandas offers a wide range of functions for data manipulation within DataFrames. Some common functions include:

.head() and tail(): These display the first or last few rows of the DataFrame, respectively. Useful for a quick view of the data.

.info() and describe(): info() provides a concise summary of the DataFrame, including data types and missing values, while describe() gives statistical
information like mean, median, and quartiles for numerical columns.

.loc[] and iloc[]: Used for accessing specific rows or columns by label (loc[]) or by integer index (iloc[]).

.drop() and dropna(): drop() removes specified rows or columns, while dropna() removes rows with missing values.

.fillna(): Fills missing values in a DataFrame with a specified value or method.

.groupby(): Groups data based on a categorical variable, allowing aggregation operations like sum, mean, count, etc., on these groups.

.apply() and applymap(): apply() executes a function along an axis of the DataFrame, and applymap() applies a function to every element of the DataFrame.

For example, you might use the groupby() function when analyzing sales data to group sales by region and then find the total sales in each region. This function
helps aggregate data based on categorical values, providing insights into regional performance or trends.





```

```python
Q6. Which of the following is mutable in nature Series, DataFrame, Panel?
Ans6. Among Series, DataFrame, and Panel in Pandas, both Series and DataFrame are mutable, meaning
they can be modified after creation.

Series: A Pandas Series is mutable. Elements within a Series can be modified, added, or removed after creation.

DataFrame: Similarly, a Pandas DataFrame is mutable. You can modify existing columns, add new columns, delete columns,
and perform various manipulations on the DataFrame after it's created.

However, Panels have been deprecated in recent versions of Pandas (as of my last update in January 2022) due to decreased usage
and complexity. Panels were three-dimensional data structures, and they have been removed from the core Pandas library. Therefore,
Panels were mutable, but their usage is no longer recommended. For most use cases, DataFrames are preferred due to their flexibility
and ease of use in handling tabular data.



```

```python
#Q7. Create a DataFrame using multiple Series. Explain with an example.

#Ans7.

import pandas as pd

# Creating multiple Pandas Series
names = pd.Series(['Alice', 'Bob', 'Charlie'])
ages = pd.Series([25, 30, 35])
cities = pd.Series(['New York', 'San Francisco', 'Los Angeles'])

# Creating a DataFrame using the Series
data = {
    'Name': names,
    'Age': ages,
    'City': cities
}

df = pd.DataFrame(data)

print(df)
```
