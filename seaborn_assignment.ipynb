```python
Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.
Ans1. Certainly! Seaborn is a powerful library for creating various statistical visualizations. Here are five types of plots and their common uses:

Scatter Plot
Use: Scatter plots are great for visualizing the relationship between two continuous variables. They help in identifying patterns,
clusters, or trends within the data.

Histogram
Use: Histograms display the distribution of a single numerical variable. They're useful for understanding the underlying frequency or
density of data points within specific ranges or bins.

Bar Plot
Use: Bar plots are ideal for comparing categorical data. They display the relationship between a categorical variable and a numerical variable,
helping to visualize comparisons among different categories.

Box Plot
Use: Box plots (or box-and-whisker plots) are excellent for displaying the distribution of data and identifying outliers.
They show the median, quartiles, and potential outliers in a dataset.

Heatmap
Use: Heatmaps are used to visualize matrix-like data where values are represented by colors. They're great for showcasing correlations, patterns,
or relationships among different variables in a dataset.

Each of these plots serves a specific purpose in visualizing data and helps in gaining insights or understanding relationships within the dataset.

```

```python
#Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
#"timepoint" and y = "signal" for different events and regions.

#Ans2
import seaborn as sns
import matplotlib.pyplot as plt
fmri= sns.load_dataset("fmri")
sns.lineplot(x= fmri.timepoint, y= fmri.signal,data = fmri, hue= "event", style= "region")
plt.show()
```

```python
#Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
#'pclass', y = 'age' and y = 'fare'.

titan= sns.load_dataset('titanic')
plt.figure(figsize=(6,4))
sns.boxplot(x= titan.pclass, y= titan.age)

plt.figure(figsize=(6,4))
sns.boxplot(x= titan.pclass, y = titan.fare)
```

```python
#Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
#parameter for the 'cut' column of the diamonds dataset.
dia= sns.load_dataset("diamonds")
sns.histplot(x=dia.price, data= dia, hue= "cut" )
```

```python
#Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
# of the iris dataset.
iris= sns.load_dataset("iris")
sns.pairplot(iris, hue= "species")
```

```python
#Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

flights_data = sns.load_dataset("flights")

# Reshape the data for heatmap (pivot)
flights_pivot = flights_data.pivot("month", "year", "passengers")

# Plotting a heatmap
plt.figure(figsize=(10, 8))  # Set the figure size

# Create a heatmap using Seaborn
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="YlGnBu")

# Set plot labels and title
plt.xlabel("Year")
plt.ylabel("Month")
plt.title("Passenger Numbers over Years and Months")

# Show the plot
plt.tight_layout()
plt.show()
```
